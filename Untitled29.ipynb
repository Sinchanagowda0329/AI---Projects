```python

# =========================
# 1. INSTALL (SAFE VERSIONS)
# =========================
!pip install -q faiss-cpu pypdf sentence-transformers transformers torch

# =========================
# 2. IMPORTS
# =========================
from google.colab import files
from pypdf import PdfReader
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from transformers import pipeline
import re

# =========================
# 3. UPLOAD PDF
# =========================
uploaded = files.upload()
pdf_file = list(uploaded.keys())[0]

print("Using:", pdf_file)

# =========================
# 4. READ PDF
# =========================
reader = PdfReader(pdf_file)

text = ""
for page in reader.pages:
    if page.extract_text():
        text += page.extract_text() + "\n"

print("PDF loaded ✔")

# =========================
# 5. CHUNK TEXT (SAFE LIMIT)
# =========================
chunks = re.split(r'\n\s*\n', text)
chunks = [c.strip() for c in chunks if len(c.strip()) > 50]
chunks = chunks[:120]   # prevent freezing

print("Chunks:", len(chunks))

# =========================
# 6. EMBEDDINGS MODEL
# =========================
model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(chunks, show_progress_bar=False)

# =========================
# 7. FAISS INDEX
# =========================
index = faiss.IndexFlatL2(len(embeddings[0]))
index.add(np.array(embeddings))

print("Vector DB ready ✔")

# =========================
# 8. LLM (NO KEYERROR FIX)
# =========================
llm = pipeline(
    "text-generation",
    model="google/flan-t5-small"
)

print("\n🚀 PDF CHATBOT READY\n")

# =========================
# 9. CHAT LOOP
# =========================
while True:
    query = input("Ask: ")

    if query.lower() == "exit":
        break

    # retrieve top chunks
    q_vec = model.encode([query])
    _, I = index.search(np.array(q_vec), k=3)

    context = "\n".join([chunks[i] for i in I[0]])

    prompt = f"""
Answer ONLY using the context.

Context:
{context}

Question:
{query}

Answer:
"""

    result = llm(prompt, max_new_tokens=120, do_sample=False)

    print("\nAnswer:\n")
    print(result[0]["generated_text"])
    print("\n" + "-"*50)
```
